## KNN using iris dataset

### Divide data into features and label

```python
X=data.iloc[:,0:4].values # features
Y=data.iloc[:,4].values # Labels
Y
```

### Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
Y=LE.fit_transform(Y)
Y
```

### Splitting dataset into train and test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.33)
print(x_train)
```

### Create and train Model

```python
from sklearn.neighbors import KNeighborsClassifier

classifier=KNeighborsClassifier(n_neighbors=3)

classifier.fit(x_train,y_train)
```

### Make prediction with model

```python
y_pred=classifier.predict(x_test)
print(y_test)
print(y_pred)
```

### Evaluate the predictor

```python
from sklearn.metrics import confusion_matrix, accuracy_score

cm=confusion_matrix(y_test,y_pred)
cm

print("Accuracy of the model is ",accuracy_score(y_test,y_pred)*100,"%")
```

### Accuracy and the error rate 

```python
classifier=KNeighborsClassifier(n_neighbors=9)
classifier.fit(x_train,y_train)
y_pred=classifier.predict(x_test)
print("Accuracy of the model is ",accuracy_score(y_test,y_pred),"%")
print("Error rate of the model is ",round(1-accuracy_score(y_test,y_pred),2),"%")
```
